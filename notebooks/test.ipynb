```python
import pandas as pd
```

```python
df = pd.read_excel("../data/DataSet.xlsx")
```

```python
df
```

```python
df = df.dropna()
```

```python
all_keywords = {}
for keywords in df['Từ Khóa']:
    for kw in keywords.split(','):
        kw = kw.strip(" ,\'\"")
        if len(kw) > 2:
            all_keywords[kw] = all_keywords.get(kw, 0) + 1
```

```python
all_keywords
```

```python
dict(sorted(all_keywords.items(), key=lambda item: item[1], reverse=True))
```

```python
import matplotlib.pyplot as plt

# Filter: only occurrences > 7
filtered_counts = {k: v for k, v in all_keywords.items() if v >= 10}

# Sort (optional, for readability)
filtered_counts = dict(sorted(filtered_counts.items(), key=lambda x: x[1], reverse=True))

# Extract keys and values
keys = list(filtered_counts.keys())
values = list(filtered_counts.values())

# Plot
plt.figure(figsize=(8, 5))
plt.bar(keys, values)
plt.xlabel('Items')
plt.ylabel('Occurrences')
plt.title('Items with Occurrences >= 10')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Clean dataframe: keep only locations with keywords in filtered_counts
def filter_keywords(keywords_str):
    """Keep only keywords that are in filtered_counts"""
    keywords_list = []
    for kw in keywords_str.split(','):
        kw = kw.strip(" ,\'\"")
        if kw in filtered_counts:
            keywords_list.append(kw)
    return ', '.join(keywords_list) if keywords_list else None

# Apply filter to create new column
df['Từ Khóa Filtered'] = df['Từ Khóa'].apply(filter_keywords)
df['Rating'] = df['Đánh giá'].str.replace('/5', '').astype(float) / 5
df['Rating'].fillna(df['Rating'].mean(), inplace=True)

# Keep only rows that have at least one keyword in filtered_counts
df_cleaned = df[df['Từ Khóa Filtered'].notna()].copy()

# Replace the original keyword column with the filtered one
df_cleaned['Từ Khóa'] = df_cleaned['Từ Khóa Filtered']
df_cleaned = df_cleaned.drop(columns=['Từ Khóa Filtered', 'Mô tả', 'Vị trí', 'Ảnh', 'STT', 'Đánh giá']).rename(columns={'Tên địa điểm': 'Location Name', 'Từ Khóa': 'Keywords'})

print(f"Original dataframe shape: {df.shape}")
print(f"Cleaned dataframe shape: {df_cleaned.shape}")
print(f"Rows removed: {df.shape[0] - df_cleaned.shape[0]}")

df_cleaned
```

```python
df_cleaned.to_csv("../data/final_dataset.csv", index=False)
```
